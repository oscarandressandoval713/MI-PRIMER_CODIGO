```python
# --- CALCULADORA INTERACTIVA PARA GOOGLE COLAB ---
# Este programa utiliza ipywidgets para crear una interfaz gráfica de calculadora
# con un teclado numérico y una pantalla, similar a una calculadora real.

import ipywidgets as widgets
from IPython.display import display, clear_output

# --- Estado de la Calculadora ---
# Variables para guardar el estado de la operación
calc_state = {
    "input": "",
    "memoria": 0,
    "operador": None,
    "reset_input": False
}

# --- Creación de los Widgets ---

# Pantalla de la calculadora
pantalla = widgets.Text(value="", description="", layout=widgets.Layout(width='260px'), disabled=True)

# Definición de los botones
button_layout = widgets.Layout(width='60px', height='40px')
buttons = {
    '7': widgets.Button(description='7', layout=button_layout),
    '8': widgets.Button(description='8', layout=button_layout),
    '9': widgets.Button(description='9', layout=button_layout),
    '÷': widgets.Button(description='÷', layout=button_layout, button_style='info'),
    '4': widgets.Button(description='4', layout=button_layout),
    '5': widgets.Button(description='5', layout=button_layout),
    '6': widgets.Button(description='6', layout=button_layout),
    '×': widgets.Button(description='×', layout=button_layout, button_style='info'),
    '1': widgets.Button(description='1', layout=button_layout),
    '2': widgets.Button(description='2', layout=button_layout),
    '3': widgets.Button(description='3', layout=button_layout),
    '-': widgets.Button(description='-', layout=button_layout, button_style='info'),
    '0': widgets.Button(description='0', layout=button_layout),
    '.': widgets.Button(description='.', layout=button_layout),
    '=': widgets.Button(description='=', layout=button_layout, button_style='success'),
    '+': widgets.Button(description='+', layout=button_layout, button_style='info'),
    'C': widgets.Button(description='C', layout=widgets.Layout(width='260px', height='40px'), button_style='danger')
}

# --- Lógica de los Eventos ---

def on_button_click(b):
    """Manejador de eventos para todos los botones."""
    char = b.description

    if char.isdigit() or char == '.':
        # Si se debe limpiar la pantalla antes de escribir, lo hacemos
        if calc_state["reset_input"]:
            calc_state["input"] = ""
            calc_state["reset_input"] = False
        # Evitar múltiples puntos decimales
        if char == '.' and '.' in calc_state["input"]:
            return
        calc_state["input"] += char

    elif char in ['+', '-', '×', '÷']:
        # Si ya hay un operador, primero calcula el resultado parcial
        if calc_state["operador"] and calc_state["input"]:
            on_button_click(buttons['=']) # Simula presionar '='

        if calc_state["input"]:
            calc_state["memoria"] = float(calc_state["input"])

        calc_state["operador"] = char
        calc_state["reset_input"] = True # Prepara para el siguiente número

    elif char == '=':
        if not calc_state["operador"] or not calc_state["input"]:
            return

        num2 = float(calc_state["input"])
        num1 = calc_state["memoria"]
        op = calc_state["operador"]
        resultado = 0

        try:
            if op == '+': resultado = num1 + num2
            elif op == '-': resultado = num1 - num2
            elif op == '×': resultado = num1 * num2
            elif op == '÷':
                if num2 == 0:
                    calc_state["input"] = "Error"
                    pantalla.value = calc_state["input"]
                    return
                resultado = num1 / num2

            # Formatear resultado para no mostrar '.0' en enteros
            calc_state["input"] = str(int(resultado) if resultado.is_integer() else round(resultado, 6))
            calc_state["memoria"] = resultado
            calc_state["operador"] = None
            calc_state["reset_input"] = True

        except Exception:
            calc_state["input"] = "Error"

    elif char == 'C':
        # Limpiar todo el estado
        calc_state["input"] = ""
        calc_state["memoria"] = 0
        calc_state["operador"] = None
        calc_state["reset_input"] = False

    # Actualizar la pantalla en cada evento
    pantalla.value = calc_state["input"]

# Asignar el manejador de eventos a cada botón
for btn in buttons.values():
    btn.on_click(on_button_click)

# --- Ensamblaje de la Interfaz ---

# Organizar los botones en filas (HBox)
fila1 = widgets.HBox([buttons['7'], buttons['8'], buttons['9'], buttons['÷']])
fila2 = widgets.HBox([buttons['4'], buttons['5'], buttons['6'], buttons['×']])
fila3 = widgets.HBox([buttons['1'], buttons['2'], buttons['3'], buttons['-']])
fila4 = widgets.HBox([buttons['0'], buttons['.'], buttons['='], buttons['+']])

# Título y contenedor principal (VBox)
titulo = widgets.HTML("<h2>Calculadora Estilo Real</h2>")
calculadora_gui = widgets.VBox([
    titulo,
    pantalla,
    fila1,
    fila2,
    fila3,
    fila4,
    buttons['C']
])

# --- Mostrar la Calculadora ---
display(calculadora_gui)



```
